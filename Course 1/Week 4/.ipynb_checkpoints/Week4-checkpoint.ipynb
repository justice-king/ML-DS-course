Логнормальное распределение

Дисперсия = $(e^{{\sigma ^{2}}}\!\!-1)e^{{2\mu +\sigma ^{2}}}$

Математическое ожидание = $e^{{\mu +\sigma ^{2}/2}}$

```python
import math
import numpy as np
import scipy
import matplotlib
from numpy import random
from matplotlib import pyplot as plt

import scipy.stats as sts
from scipy.stats import lognorm
```

$\sigma = 1$

$\mu = 1$

Генерируем выборку размера 1000 из логнормального распределения

```python
def lognorm(sigma, mu, x):
    return (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi)))
mu = 1.0
sigma = 1.0

sample = [np.random.lognormal(mu, sigma) for i in range(1000)]
```

Гистограмма на нашей выборке и функция плотности вероятности нашего распределения

```python
plt.hist(sample, bins = 100, normed = True)

X = np.arange(0.1, 80, 0.05)
Y = np.array([lognorm(mu, sigma, i) for i in X])

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(X, Y)
```

Среднее арифметическое из объема выборки размера size

```python
def getsample(size):
    res = 0.0
    for i in range(size):
        res += np.random.lognormal(mu, sigma)
    return res / size
```

Гистограмма на выборке размера 1000 с объемом выборки 5 и функция нормального распределения с параметрами мат ожидание нашего распределения и дисперсией нашего распределение деленной на объем нашей выборки
$(e^{{\sigma ^{2}}}\!\!-1)e^{{2\mu +\sigma ^{2}}}$
и $e^{{\mu +\sigma ^{2}/2}}/5$

```python
samplesizeof5 = [getsample(5) for i in range(1000)]
plt.hist(samplesizeof5, bins = 100, normed = True)

dispersion = (math.exp(sigma**2) - 1) * math.exp(2 * mu + sigma ** 2)
expect = math.exp(mu + sigma**2 / 2)
norm_rv = sts.norm(expect, math.sqrt(dispersion / 5))

x1 = np.linspace(0, 20, 100)
y1 = norm_rv.pdf(x1)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(x1, y1)
```

Гистограмма на выборке размера 1000 с объемом выборки 10 и функция нормального распределения с параметрами мат ожидание нашего распределения и дисперсией нашего распределение деленной на объем нашей выборки
$(e^{{\sigma ^{2}}}\!\!-1)e^{{2\mu +\sigma ^{2}}}$
и $e^{{\mu +\sigma ^{2}/2}}/10$

```python
samplesizeof10 = [getsample(10) for i in range(1000)]
plt.hist(samplesizeof10, bins = 100, normed = True)

dispersion = (math.exp(sigma**2) - 1) * math.exp(2 * mu + sigma ** 2)
expect = math.exp(mu + sigma**2 / 2)
norm_rv = sts.norm(expect, math.sqrt(dispersion / 10))

x1 = np.linspace(0, 20, 100)
y1 = norm_rv.pdf(x1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(x1, y1)
```

Гистограмма на выборке размера 1000 с объемом выборки 50 и функция нормального распределения с параметрами мат ожидание нашего распределения и дисперсией нашего распределение деленной на объем нашей выборки
$(e^{{\sigma ^{2}}}\!\!-1)e^{{2\mu +\sigma ^{2}}}$
и $e^{{\mu +\sigma ^{2}/2}}/50$

```python
samplesizeof50 = [getsample(50) for i in range(1000)]
plt.hist(samplesizeof50, bins = 100, normed = True)

dispersion = (math.exp(sigma**2) - 1) * math.exp(2 * mu + sigma ** 2)
expect = math.exp(mu + sigma**2 / 2)
norm_rv = sts.norm(expect, math.sqrt(dispersion / 50))

x1 = np.linspace(0, 20, 100)
y1 = norm_rv.pdf(x1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(x1, y1)
```

Чем больше объем нашей выборки тем лучше нормальное распределение описывает нашу выборку
